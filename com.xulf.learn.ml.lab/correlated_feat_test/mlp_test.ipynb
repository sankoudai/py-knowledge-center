```python
import tensorflow as tf
import matplotlib
import matplotlib.pyplot as plt
```

## data

```python
tf.random.set_seed(100)
size = 1000

x1 = tf.random.uniform([size],  minval=0, maxval=1, seed=100)

# x2 与 x1正相关
raw_x2 = tf.math.exp(0.2*x1**2) * 10 - 10
noise_x2 = 0.2 + tf.random.normal([size], mean=0., stddev=0.7, seed=100)
x2 = raw_x2 + noise_x2
# x2 = raw_x2


noise_y = 0
y = tf.sigmoid(10*x1 - 2 *x2 + noise_y)

feats  = {
        'x1': x1,
        'x2': x2
}
ds = tf.data.Dataset.from_tensor_slices((feats, y))

plt.scatter(x1, x2)
```

```python
plt.scatter(x1, y)
```

## model

- mlp可以学习相关联的特征
- 与较弱的特征，其真实相关性， 需要更多的epoch才能学会
- 所以一个问题是， 是否可以采用coordinated descent方式来学习呢？

- 下述实验中， 如果epochs=10，则多次实验才能得到y 与 x2负相关; 如果epochs=100，基本都能得到这个结论
- 如果y与x2的系数绝对值较大， epochs=10, 也基本能得到y与x2的负相关结果

```python
train_ds = ds.batch(10)

x1 = tf.keras.Input(shape=(1,), name="x1")
x2 = tf.keras.Input(shape=(1,), name="x2")
all_feats = tf.keras.layers.concatenate([x1, x2])


mlp_layer = tf.keras.Sequential([
            tf.keras.layers.Dense(8, activation='relu'),
            tf.keras.layers.Dense(1, activation='sigmoid')
        ])
output = mlp_layer(all_feats)

model = tf.keras.Model([x1, x2], output)
model.compile(
        loss=tf.losses.BinaryCrossentropy(from_logits=False), 
        optimizer='adam',
        metrics=tf.metrics.BinaryAccuracy(threshold=0.0))
model.fit(train_ds, epochs=10)

new_feats = {}
new_feats['x1'] = feats['x1']
new_feats['x2'] = feats['x2'] + 0.3
plt.scatter(feats['x2'], model(new_feats) - model(feats))

new_feats = {}
new_feats['x2'] = feats['x2']
new_feats['x1'] = feats['x1'] + 0.3
plt.scatter(feats['x1'], model(new_feats) - model(feats), marker='+')
```

```python
train_ds = ds.batch(10)

x1 = tf.keras.Input(shape=(1,), name="x1")
x2 = tf.keras.Input(shape=(1,), name="x2")
all_feats = tf.keras.layers.concatenate([x1, x2])


mlp_layer = tf.keras.Sequential([
            tf.keras.layers.Dense(8, activation='relu'),
            tf.keras.layers.Dense(1, activation='sigmoid')
        ])
output = mlp_layer(all_feats)

model = tf.keras.Model([x1, x2], output)
model.compile(
        loss=tf.losses.BinaryCrossentropy(from_logits=False), 
        optimizer='adam',
        metrics=tf.metrics.BinaryAccuracy(threshold=0.0))
model.fit(train_ds, epochs=30)

new_feats = {}
new_feats['x1'] = feats['x1']
new_feats['x2'] = feats['x2'] + 0.3
plt.scatter(feats['x2'], model(new_feats) - model(feats))

new_feats = {}
new_feats['x2'] = feats['x2']
new_feats['x1'] = feats['x1'] + 0.3
plt.scatter(feats['x1'], model(new_feats) - model(feats), marker='+')
```

##  更精确的模型

```python
train_ds = ds.batch(10)

x1 = tf.keras.Input(shape=(1,), name="x1")
x2 = tf.keras.Input(shape=(1,), name="x2")

all_feats = tf.keras.layers.concatenate([x1, x2])
mlp_layer = tf.keras.Sequential([
#             tf.keras.layers.Dense(1, activation='relu'),
            tf.keras.layers.Dense(1, activation='sigmoid')
        ])

output = mlp_layer(all_feats)

model = tf.keras.Model([x1, x2], output)
model.compile(
        loss=tf.losses.BinaryCrossentropy(from_logits=False), 
        optimizer='sgd',
        metrics=tf.metrics.BinaryAccuracy(threshold=0.0))
model.fit(train_ds, epochs=10)
```

```python
new_feats = {}
new_feats['x1'] = feats['x1']
new_feats['x2'] = feats['x2'] + 0.3
plt.scatter(feats['x2'], model(new_feats) - model(feats))

new_feats = {}
new_feats['x2'] = feats['x2']
new_feats['x1'] = feats['x1'] + 0.3
plt.scatter(feats['x1'], model(new_feats) - model(feats), marker='+')
```
